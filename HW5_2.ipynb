```python
# let's see unique values for each column
s = (data.dtypes == 'object')
categorical_cols = list(s[s].index)

for col in categorical_cols:
  print(f'{col} => {data[col].unique()}')
  print("="*50)
  print("\n")
```

As we can see in data, we have `Education` and `Education-num` columns. first one is categorical feature and the other one is numerical. let's check them.

```python
print('Education num =>',  data['Education-num'].unique())

if (len(data['Education-num'].unique()) == len(data['Education'].unique())):
  print(f'Education and Education-num have same length')

print('Education =>\n' ,data['Education'].value_counts())
print('Education-num =>\n' ,data['Education-num'].value_counts())
```

### Drop Redundant features

As we can see in the above cell, `Education` and `Education-num` are equivalent. we should drop one of them to not have redundant features. we drop `Education` column.


```python
data.drop('Education', inplace=True, axis=1)
print(f'data shape : {data.shape}')
```

```python
data.head()
```

### Split X and y

```python
X = data.drop(columns='Income')
y = pd.DataFrame(data['Income'])

print(f'X shape : {X.shape}')
print(f'y shape : {y.shape}')
```

### Encoding Categorical features

let's encoding features.

```python
X = pd.get_dummies(X)
X.head()
```

```python
print(f'X shape after encoding : {X.shape}')
```

let's encoding label of classes.


```python
le = LabelEncoder()
le.fit(y)
y = le.transform(y)
```

```python
print(f'y shape after label encoding: {y.shape}')
```

```python
print(f'{le.inverse_transform([0, 1])[0]} mapped to 0 and {le.inverse_transform([0, 1])[1]} mapped to 1')
```

## Classification using Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

lr = LogisticRegression(solver='lbfgs', random_state=42)
scores = cross_val_score(lr, X, y, cv=10)

print(f'accuracy of lr model : {scores.mean()}')
```

## Classification using Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
scores = cross_val_score(gnb, X, y, cv=10)
print(f'accuracy of gnb model : {scores.mean()}')
```
